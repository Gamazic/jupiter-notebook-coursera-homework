# нужно для каждого слова составить количество его вхождений для каждого предложений, получить лист из словарей

## fcats - файл с предложениями
## sCats -строка со всеми предложениями

```python
def Tokenize(string):
    string=string.lower()
    listOfToken=re.split('[^a-z]', string)
    listOfToken=[el for el in listOfToken if el!='']
    return listOfToken
```

```python
listOfToken=list()

fcats.seek(0)
for line in fcats:
    line=line.lower()
    listOfToken.append(Tokenize(line))
##токенезировали каждое предложение и вложили каждый лист токенов в общий лист.
print listOfToken
```
